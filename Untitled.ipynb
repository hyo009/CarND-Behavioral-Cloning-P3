```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn

import os
import csv
import sklearn
from sklearn.model_selection import train_test_split
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Lambda, Flatten, Dense, Convolution2D
from keras.models import load_model
from keras import callbacks
from keras.optimizers import Adam
%matplotlib inline
```

```python
# load data
dataLocation = './data/'
samples = []
with open(dataLocation + 'driving_log.csv') as csvfile:
    reader = csv.reader(csvfile)
    for line in reader:
        samples.append(line)
```

```python
# split training and test data
train_samples, test_samples = train_test_split(samples, test_size = 0.2)

print('training examples:', len(train_samples))
print('test examples:', len(test_samples))
```

```python
correction = 0.2 # this is a parameter to tune
```

```python
# visualize one sample
m = np.random.randint(0,len(samples))
print('training example m :', m)

sample = samples[m]
print(sample[0])
center_img = cv2.imread(dataLocation + sample[0])
left_img = cv2.imread(dataLocation + sample[1])
right_img = cv2.imread(dataLocation + sample[2])

center_angle = float(sample[3])
left_angle = center_angle + correction
right_angle = center_angle - correction
print('center_angle' , center_angle)

plt.figure
plt.imshow(center_img)

print(center_img.shape)
```

```python
# Crop top and bottom images to remove sky and front of the car (NVIDIA's paper)
def preprocessing(image):
    # crop to 90x320x3
    image = image[50:140,:,:]
    
    # scale to 66x200x3 (same as NVIDIA)
    image = cv2.resize(image,(200, 66), interpolation = cv2.INTER_AREA)

    # convert to YUV color space (same as NVIDIA)
    #return cv2.cvtColor(image, cv2.COLOR_BGR2YUV)
    
    return image
```

```python
center_img = preprocessing(center_img)

plt.figure
plt.imshow(center_img)

print(center_img.shape)
```

```python
# generators
def generator_training(samples, batch_size=128):
    while 1: # Loop forever so the generator never terminates
        sklearn.utils.shuffle(samples)
        for offset in range(0, len(samples), batch_size):
            batch_samples = samples[offset:offset+batch_size]

            images = []
            angles = []
            for batch_sample in batch_samples:
                center_image = cv2.imread(dataLocation + batch_sample[0])
                # left_image = cv2.imread(dataLocation + batch_sample[1])
                # right_image = cv2.imread(dataLocation + batch_sample[2])
                
                center_angle = float(batch_sample[3])
                # left_angle = center_angle + correction
                # right_angle = center_angle - correction
                
                # Randomly flips the images to help create even amount of left turns and right turns
                if np.random.rand() > .5:
                    center_image = cv2.flip(center_image, 1)
                    center_angle *= -1
                
                # Preprocess data            
                images.append(preprocessing(center_image))
  
                angles.append(center_angle)

            X_train = np.array(images)
            y_train = np.array(angles)
            yield sklearn.utils.shuffle(X_train, y_train)
```

```python
def generator_test(samples, batch_size=128):
    while 1: # Loop forever so the generator never terminates
        sklearn.utils.shuffle(samples)
        for offset in range(0, len(samples), batch_size):
            batch_samples = samples[offset:offset+batch_size]

            images = []
            angles = []
            for batch_sample in batch_samples:
                center_image = cv2.imread(dataLocation + batch_sample[0])
                
                center_angle = float(batch_sample[3])
                
                images.append(preprocessing(center_image))
                
                angles.append(center_angle)

            X_train = np.array(images)
            y_train = np.array(angles)
            yield sklearn.utils.shuffle(X_train, y_train)
```

```python
# compile and train the model using the generator function
train_generator = generator_training(train_samples, batch_size = 256)
validation_generator = generator_test(test_samples, batch_size = 256)
```

```python
def nvidia_model():   
    model = Sequential()
    
    model.add(Lambda(lambda x: x/127.5 - 1, input_shape = (66,200,3)))
    
    model.add(Convolution2D(24, 5, 5, subsample = (2, 2), border_mode = "valid", activation = 'elu'))
    model.add(Convolution2D(36, 5, 5, subsample = (2, 2), border_mode = "valid", activation = 'elu'))
    model.add(Convolution2D(48, 5, 5, subsample = (2, 2), border_mode = "valid", activation = 'elu'))
    model.add(Convolution2D(64, 3, 3, subsample = (1, 1), border_mode = "valid", activation = 'elu'))
    model.add(Convolution2D(64, 3, 3, subsample = (1, 1), border_mode = "valid", activation = 'elu'))
    
    model.add(Flatten())
    
    model.add(Dense(100, activation = 'elu'))
    model.add(Dense(50, activation = 'elu'))
    model.add(Dense(10, activation = 'elu'))
    model.add(Dense(1, activation = 'elu'))
    
    model.summary()
    
    return model
```

```python
# create a new model
model = nvidia_model()
```

```python
# train the model
model.compile(loss='mse', optimizer='adam')
model.fit_generator(train_generator, 
                    samples_per_epoch=len(train_samples), 
                    validation_data=validation_generator,
                    nb_val_samples=len(test_samples), 
                    nb_epoch=3)
```
